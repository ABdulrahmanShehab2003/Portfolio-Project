```python
# Create a scatter chart of 'budget' vs 'gross'

plt.scatter(x = movie['budget'], y = movie['gross'])

# Set chart title and labels
plt.title('Budget vs Gross')
plt.xlabel('Budget')
plt.ylabel('Gross')

plt.figure(figsize=(20,10))
# Show the chart
plt.show()
```

```python
#plot the budget vs gross using seaborn to see the line

sns.regplot(data=movie, x='budget', y='gross' ,scatter_kws={"color" :"r"} , line_kws={"color":"blue"})
```

```python
#start looking at correlation
movie.corr(method='pearson')
```

```python
sns.heatmap(movie.corr(method='pearson'), annot=True)
plt.title('Correlation Matrix for Numeric Features')
plt.show()
```

```python
ax = movie[['genre','votes']].groupby(['genre']).sum().sort_values(by=['votes'], ascending=True)\
.plot(kind='bar', title = 'Most Popular Movie Genres by Voting Numbers')
ax.set_xlabel('Genres') #labeling the x axis
ax.set_ylabel('Votes')

plt.show()
```
